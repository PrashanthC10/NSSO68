# 5. Test whether the differences in the means are significant or not.

```python
stats.ttest_ind(mh.MPCE_MRP,mh.MPCE_URP)
```

```python
mh['Sector'].value_counts()
```

```python
mh[mh['Sector'] == 'Rural']['MPCE_MRP'].mean()
```

```python
mh[mh['Sector'] == 'Urban']['MPCE_MRP'].mean()
```

```python
rural = mh[mh['Sector'] == 'Rural']['MPCE_MRP']
urban = mh[mh['Sector'] == 'Urban']['MPCE_MRP']
```

```python
len(rural), len(urban)
```

```python
stats.ttest_ind(rural, urban)
```

### If the p-value is inferior or equal to the significance level 0.05, we can reject the null hypothesis and accept the alternative hypothesis. In other words, we conclude that the sample mean is significantly different from the theoretical mean.

```python
from statsmodels.stats.weightstats import ztest
ztest(rural,urban)
```

### A positive z-score indicates the raw score is higher than the mean average. A negative z-score reveals the raw score is below the mean average.
